#Data Cleaning and EDA on Heart Disease Dataset
This notebook walks through data cleaning and exploratory data analysis (EDA) on a heart disease dataset. We'll handle missing values, inspect data types, and visualize key patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

##Load the Dataset

```python
df = pd.read_csv('heart.csv')
df.head()
```

##Basic Info and Summary Stats

```python
df.info()
```

```python
df.describe(include='all')
```

## 🔍 Check for Missing Values

```python
df.isnull().sum()
```

##Data Cleaning: Check Data Types

```python
df.dtypes
```

##Remove Duplicates

```python
df = df.drop_duplicates()
df.shape
```

##Visualize Distributions of Numerical Features

```python
num_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
df[num_cols].hist(figsize=(12, 8), bins=20, color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

##Distribution of Categorical Features

```python
cat_cols = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
for col in cat_cols:
    sns.countplot(data=df, x=col)
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

##Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

##Target Variable vs Features

```python
sns.countplot(x='HeartDisease', data=df)
plt.title('Target Variable Distribution')
plt.show()
```

##Summary
- No missing values were found in the dataset.
- Some duplicates were removed.
- Data types are appropriate.
- Key distributions and correlations were visualized.

This dataset is now ready for feature engineering or modeling.
